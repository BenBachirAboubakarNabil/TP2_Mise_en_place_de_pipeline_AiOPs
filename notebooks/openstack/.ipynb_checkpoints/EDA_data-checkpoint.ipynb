```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import sys
sys.path.append('/app/notebooks/openstack')
from utils import *
```

#### Charger données prétraitées

```python
df = load_data()

print(f"Shape: {df.shape}")
print(df.head())
```

#### Exploration des donnees

```python
# Affichez les types de données de toutes les colonnes 
print(df.info()) 

missing_counts = df.isnull().sum()
missing_percent = 100 * missing_counts / len(df) # Calcul du pourcentage

missing_df = pd.DataFrame({
    'Manquantes': missing_counts,
    'Pourcentage': missing_percent
})

# Afficher uniquement les colonnes qui ont des valeurs manquantes et trier par pourcentage
missing_to_show = missing_df[missing_df['Manquantes'] > 0].sort_values(
    by='Pourcentage', ascending=False
)
print(missing_to_show)
```

### Analyse de la distribution et de la relation des caractéristiques

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Barplot
df['Label'].value_counts().plot(kind='bar', ax=ax1, color=['skyblue', 'salmon'])
ax1.set_title('Distribution des Labels', fontweight='bold', fontsize=14)
ax1.set_xlabel('Label', fontsize=12)
ax1.set_ylabel('Nombre d\'instances', fontsize=12)
ax1.grid(axis='y', alpha=0.3)

# Pie chart
df['Label'].value_counts().plot(kind='pie', ax=ax2, autopct='%1.1f%%', 
                                colors=['skyblue', 'salmon'], startangle=90)
ax2.set_ylabel('')
ax2.set_title('Proportion des Labels', fontweight='bold', fontsize=14)

plt.tight_layout()
plt.show()

```

```python
event_cols = get_event_columns(df)
print(f"\nNombre d'evenements: {len(event_cols)}")

# Total occurrences par événement
event_totals = df[event_cols].sum().sort_values(ascending=False)
print(f"\nTop 15 evenements les plus frequents:")
print(event_totals.head(15))
```

```python
top_20_events = event_totals.head(20).index.tolist()

plt.figure(figsize=(15, 12))
corr = df[top_20_events].corr()
sns.heatmap(corr, annot=True, cmap='viridis', center=0, fmt=".2f", linewidths=.5)
plt.title('Matrice de correlation d\'evenements')
plt.show()


high_corr_pairs = corr.unstack().sort_values(ascending=False)
high_corr_pairs = high_corr_pairs[(high_corr_pairs > 0.9) & (high_corr_pairs < 1.0)]

if not high_corr_pairs.empty:
    print("\nevenement fortement correle:")
    print(high_corr_pairs)
```
